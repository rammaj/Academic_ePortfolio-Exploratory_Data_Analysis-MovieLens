### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
# Data frame where the release year is the index and the genre is given by the column names
# Dropped 'movie id' because it is not required here
grouped_movies = movies.groupby(['release year']).sum().drop(['movie id', 'unknown'], axis = 1)
grouped_movies.head()
```

```python
# Find the number of movies for each year based on genre. Store that value in a new column as 'total'
# Note: Assuming the "total" on each year is the sum of genres within the year, regardless the amount of movies released on that year
# this way the total % per row will be 100%
grouped_movies['total per genre'] = grouped_movies.sum(axis = 1)
grouped_movies
```

```python
# Now divide the value of each genre in that year by "total" to get percentage number of release in a particular year. (df.div(df['total'], axis= 0) * 100)
grouped_movies = grouped_movies.div(grouped_movies['total per genre'], axis = 0) * 100
grouped_movies.drop(['total per genre'], axis = 1, inplace = True)
grouped_movies
```

```python
sns.set({'figure.figsize': (25,16)})
sns.set_context("notebook")
sns.heatmap(grouped_movies, annot = True, cmap = "Greens", fmt = '.2f', linewidths = 0.05);
```

```python
sns.set({'figure.figsize': (22, 6)})
sns.set_style("whitegrid")
sns.set_style("ticks")
sns.set_context("notebook")
g = sns.lineplot(x = "release year", y = "percentage", hue = "variable", palette = "tab20", \
             data = pd.melt(frame = grouped_movies.reset_index(), id_vars=['release year'], value_name = 'percentage'))
g.legend(loc='center right', bbox_to_anchor=(1.1, 0.5));
```

**Insights:**

<span style="font-family: Arial; font-weight:bold;font-size:1.0em;color:blue;">

2) After 1993, it's clear that Drama is the most popular genre (assuming the definition of popularity from this example is valid).

### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# Find the count of ratings and average ratings for every movie.
# Note: Here I'm assuming every 'movie id' is a unique movie.
ratings_merge = pd.merge(rating.groupby(['movie id']).count()['rating'], rating.groupby(['movie id']).mean()['rating'], on = "movie id")
ratings_merge.columns = ['ratings count', 'rating average']
ratings_merge['rating average'] = ratings_merge['rating average']
ratings_merge.head()
```

```python
# Merging "movie" and "ratings_count_average_merge" datasets
movies_rating_merge = pd.merge(movies, ratings_merge, on = 'movie id')
movies_rating_merge.head()
```

```python
# Slice the movies that have been rated more than 100 times
# Sort values according to average rating such that movie which highest rating is on top
# Select top 25 movies
top25 = movies_rating_merge[movies_rating_merge['ratings count'] > 100].sort_values(by = 'rating average', ascending = False).iloc[0:25, ]
top25.head()
```

```python
top25[['movie id', 'movie title', 'ratings count', 'rating average']]
```

<span style="font-family: Arial; font-weight:bold;font-size:1.0em;color:blue;">

### 10. Check for the validity of the below statements with respect to the data provided - 10 marks

* Men watch more Romance than women

**compare the percentages**

**Please pay attention to what should be the denominator while calculating percentages**

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
# Merging "movie", "rating" and "user" datasets
all_merged = pd.merge(pd.merge(movies, rating, on = 'movie id'), user, on = 'user id')
all_merged.head()
```

```python
all_merged.shape
# Original "rating" dataset had 100000 rows, but the merged dataset has 99990 rows
# This difference os because on the "movies" dataset there isn't a "movie id"=267 (09 rows on "rating" dataset) and also 01 row for "genre"=unknown was deleted.
```

```python
# Total amount of ratings per gender
Female_count = all_merged.groupby(['gender']).count()['rating'][0]
Male_count = all_merged.groupby(['gender']).count()['rating'][1]
print("The overall count of ratings for Female is", Female_count)
print("The overall count of ratings for Male is", Male_count)
```

```python
# Do men watch more Drama than women?
Drama_gender = all_merged[(all_merged['Drama'] == 1)].groupby(['gender']).count()['rating'].reset_index()
Drama_gender['gender total count'] = [Female_count, Male_count]
Drama_gender['Percentage'] = 100 * Drama_gender['rating'] / Drama_gender['gender total count']
Drama_gender
```

```python
sns.set({'figure.figsize': (8, 6)})
sns.set_style("whitegrid")
sns.set_style("ticks")
sns.set_context("notebook")
total = float(len(Drama_gender)) 
ax = sns.barplot(x="gender", y = "Percentage", data = Drama_gender)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2,
            height + 0.1,
            '{:1.2f}'.format(height) + "%",
            ha = "center")
ax.set_title("Drama Movies Preference - % per Gender");
```

```python
# Women watch more Sci-Fi than men?
SciFi_gender = all_merged[(all_merged['Sci-Fi'] == 1)].groupby(['gender']).count()['rating'].reset_index()
SciFi_gender['gender total count'] = [Female_count, Male_count]
SciFi_gender['Percentage'] = 100 * SciFi_gender['rating'] / SciFi_gender['gender total count']
SciFi_gender
```

```python
sns.set({'figure.figsize': (8, 6)})
sns.set_style("whitegrid")
sns.set_style("ticks")
sns.set_context("notebook")
total = float(len(SciFi_gender)) 
ax = sns.barplot(x="gender", y = "Percentage", data = SciFi_gender)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2,
            height + 0.1,
            '{:1.2f}'.format(height) + "%",
            ha = "center")
ax.set_title("Sci-Fi Movies Preference - % per Gender");
```

```python
# Do men watch more Romance than women?
Romance_gender = all_merged[(all_merged['Romance'] == 1)].groupby(['gender']).count()['rating'].reset_index()
Romance_gender['gender total count'] = [Female_count, Male_count]
Romance_gender['Percentage'] = 100 * Romance_gender['rating'] / Romance_gender['gender total count']
Romance_gender
```
